```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
dataset = pd.read_csv("uber-raw-data-sep14.csv")
dataset.head()
```

```python
dataset.shape
```

```python
dataset.info()
```

```python
dataset["Date/Time"] = dataset["Date/Time"].map(pd.to_datetime) 
dataset.head()
```

```python
dataset.dtypes
```

```python
dataset["Day"] = dataset["Date/Time"].apply(lambda x: x.day)
dataset["Hour"] = dataset["Date/Time"].apply(lambda x: x.hour)
dataset["Weekday"] = dataset["Date/Time"].apply(lambda x: x.weekday())
dataset.head(5)
```

```python
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(dataset.Day, width= 0.6, bins= 30)
plt.title("Density of trips per Day", fontsize=16)
plt.xlabel("Day", fontsize=14)
plt.ylabel("Density of rides", fontsize=14)
```

```python
#Visualize the Density of rides per Weekday
fig,ax = plt.subplots(figsize = (12,6))
plt.hist(dataset.Weekday, width= 0.6, range= (0, 6.5), bins=7, color= "green")
plt.title("Density of trips per Weekday", fontsize=16)
plt.xlabel("Weekday", fontsize=14)
plt.ylabel("Density of rides", fontsize=14)
```

```python
#Visualize the Density of rides per location
fig,ax = plt.subplots(figsize = (12,6))
x= dataset.Lon
y= dataset.Lat
plt.scatter(x, y, color= "orange")
plt.title("Density of trips per Hour", fontsize=16)
plt.xlabel("Hour", fontsize=14)
plt.ylabel("Density of rides", fontsize=14)
```
